```python
catFeature = data.columns[data.dtypes == object]
catFeature
catFeature.tolist()
```

```python
#Aplicar codificador de etiquetas

from sklearn.preprocessing import LabelEncoder,OneHotEncoder

#LabelEncoder: codifica las etiquetas de destino con un valor entre 0 y n_classes-1.
# OneHotEncoder: crea una columna binaria para cada categoría y devuelve una matriz dispersa o una matriz densa

le = LabelEncoder()
#fit_transform --> Ajusta el codificador de etiquetas y devuelva las etiquetas codificadas
x = data[catFeature].apply(le.fit_transform)


one = OneHotEncoder()
oneData = one.fit_transform(x).toarray()
oneData.shape
```

## STANDARD SCALER
La idea detrás de StandardScaler es que transformará sus datos de modo que su distribución tenga un valor medio 0 y una desviación estándar de 1.
En el caso de datos multivariados, esto se hace en función de las características (en otras palabras, independientemente para cada columna de datos).
Dada la distribución de los datos, cada valor en el conjunto de datos tendrá el valor medio restado, y luego dividido por la desviación estándar de todo el conjunto de datos (o característica en el caso multivariante).

```python

#Reescalar columnas numéricas

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#Columnas numéricas
conFeature = data.columns[data.dtypes != object]
conFeature = conFeature.drop(['SalePrice']) 
col = conFeature.tolist()


col
```

```python
#Aplicamos Standard Scaler 
dataNumeric = data[col]
sc = StandardScaler()
scaler = sc.fit(dataNumeric)
dataTransform = scaler.transform(dataNumeric)
```

```python
#Creamos dataframe con columnas numéricas y categóricas codificadas
newData = np.append(dataTransform,x,axis=1)
newData.shape
```

```python
newData
```

```python
#Separar datos de entrenamiento y de test

xTrain,xTest,yTrain,yTest = train_test_split(newData,data['SalePrice'],test_size=0.2)
```

## Árboles de decisión problema de regresión
d

```python
#Aplicar árboles de decisión 

from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(xTrain, yTrain)
```

```python
predictValues = regressor.predict(xTest)
```

```python
predictValues
```

```python
## Hallar Error cuádratico medio

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(predictValues, yTest)
mse
```

## Random Forest problema de regresión

```python
from sklearn.ensemble import RandomForestRegressor

randomForestModel = RandomForestRegressor(n_estimators=500)
randomForestModel.fit(xTrain, yTrain)
```

```python
#Predicción
predictValuesRF=randomForestModel.predict(xTest)
```

```python
mseRF = mean_squared_error(predictValuesRF, yTest)
mseRF
```

## Árboles de decisión, problema de clasificación

```python
data['SalePrice']
```

```python
groupHousing = [
    (data['SalePrice'] <= 100000),
    (data['SalePrice'] > 100000) & (data['SalePrice'] <= 500000),
    (data['SalePrice'] > 500001)]
group = [1,2,3]
```

```python
data['SalePrice_Grupo'] = np.select(groupHousing, group, default='null')
```

```python
#Separar datos de entrenamiento y de test

xTrain,xTest,yTrain,yTest = train_test_split(newData,data['SalePrice_Grupo'],test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

modelTree = DecisionTreeClassifier(random_state=0)
modelTree.fit(xTrain, yTrain)
```

```python
predictTree = modelTree.predict(xTest)
```

```python
pd.crosstab(yTest, predictTree, rownames=["Real"], colnames=["Predicho"])
```

```python
from sklearn.metrics import classification_report
print(classification_report(yTest, predictTree))
```

## Ramdom Forest clasificación

```python
from sklearn.ensemble import RandomForestClassifier

modelRFC = RandomForestClassifier(random_state=0)
modelRFC.fit(xTrain, yTrain)
```

```python
predictRFC = modelRFC.predict(xTest)
```

```python
confusionMatrixRFC = pd.crosstab(yTest, predictRFC, rownames=["Real"], colnames=["Predicho"])
confusionMatrixRFC
```

```python
print(classification_report(yTest, predictRFC))
```
